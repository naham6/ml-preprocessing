```python
plt.subplot(1,3,1)
plt.plot(x,y,'r--')
plt.subplot(1,3,2)
plt.plot(x,y,'r--')
plt.subplot(1,3,3)
plt.plot(x,y,'r--')
```

```python
fig=plt.figure()
axs=fig.add_axes([0.2,0.5,0.5,0.5])
axs.plot(x,y,'g')
axs.set_xlabel('x')
axs.set_ylabel('y')
```

```python
fig = plt.figure()
axs1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axs2 = fig.add_axes([0.1, 0.7, 0.4, 0.3])
axs1.plot(x, y, 'g')
axs1.set_xlabel('x')
axs1.set_ylabel('y')
axs1.set_title('Axes 2')
axs2.plot(y, x, 'r')
axs2.set_xlabel('x')
axs2.set_ylabel('y')
axs2.set_title('Axes 2')
```
