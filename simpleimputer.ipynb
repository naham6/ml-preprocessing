Filling values using simpleimputer

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
d1=pd.read_csv('/content/menu_items - menu_items.csv')
```

```python
d2=pd.read_csv('/content/1 - 1.csv')
```

```python
d2.head()
```

```python
d1.rename(columns={'menu_item_id':'item_id'},inplace=True)
d1.head()
```

```python
print(len(d2))
print(len(d1))
```

```python
d3=d1.merge(d2,on='item_id')
d3.head()
```

```python
d3.isnull().sum()
```

```python
d3.dtypes
```

```python
simplemode=SimpleImputer(strategy='most_frequent')
```

```python
d3[['category']]=simplemode.fit_transform(d3[['category']])
```

```python
d3.isnull().sum()
```

```python
d3[['order_date']]=simplemode.fit_transform(d3[['order_date']])
```

```python
d3[['order_time']]=simplemode.fit_transform(d3[['order_time']])
```

```python
simplemean=SimpleImputer(strategy='mean')
```

```python
d3[['price']]=simplemean.fit_transform(d3[['price']])
```

```python
d3.isnull().sum()
```
